<p style="color: blue">Newton's Law of Universal Gravitation relates the magnitude of the attractive force, $F$, between two point masses, $m_1$ and $m_2$, to the inverse square of their separation, $r$,</p>

<p style="color: blue"><a name="eq:grav">*Equation 1:*</a>  Newton's Law of Universal Graviation
\begin{equation}
  F = G \frac{m_1 m_2}{r^2},
\end{equation}</p>

<p style="color: blue">where $G$ is the gravitational constant.
It can be shown that this equation also holds for spherical masses of uniform density if $r$ is the distance between their centers.</p>

<p style="color: blue">In 1798, Henry Cavendish published the first laboratory-based direct measurement of $G$ (Cavendish, H., _"XXI. Experiments to determine the density of the earth"_, **Phil. Trans. R. Soc.**, vol. 88, p. 469, [doi:10.1098/rstl.1798.0022](https://doi.org/10.1098/rstl.1798.0022)).

<p style="text-align: center; color: blue">Figure 1:  Illustration of Cavendish's experimental apparatus, **Phil. Trans. R. Soc.**, vol. 88, p. 526, [doi:10.1098/rstl.1798.0022](https://doi.org/10.1098/rstl.1798.0022).

<p style="color: blue">The experimental apparatus is a simple torsion pendulum, similar to the one depicted in [Fig. 1](#Fig1).
Instead of swinging in a vertical plane, a torsion pendulum twists in a horizontal plane.
A horizontal rod is suspended from its center by a wire (in [Fig. 1](#Fig1), $n$, $h$, and $m$ are points on the rod).
Twisting the rod in the horizontal plane away from its rest position by a small angle $\theta$ results in a counteracting torque from the wire, $\tau$, given by</p>

<p style="color: blue"><a name="eq:thook">*Equation 2:*</a>  Torque of a torsion wire
\begin{equation}
  \tau = -\kappa\,\theta,
\end{equation}</p>

<p style="color: blue">where $\kappa$ is a constant of the wire called the torsion constant.
Note that this has the same mathematical form as Hooke's Law for a spring,</p>

<p style="color: blue"><a name="eq:hooke">*Equation 3:*</a>  Hooke's Law
\begin{equation}
  F = -k\,x,
\end{equation}</p>

<p style="color: blue">from the Simple Harmonic Motion (SHM) laboratory.
The analogy with spring motion is excellent, and the same mathematics describe the twisting of a torsion pendulum and the the translation of a mass on a spring.</p>

<a name="Fig2"></a>


<p style="text-align: center; color: blue">Figure 2:  Top view of a torsion pendulum looking downward.
Each arm of the pendulum has a length $r$ and a lead sphere of mass $m$ attached to its end.
Two large lead spheres of mass $M$ are placed in symmetric positions as indicated with their centers in the same horizontal plane as masses $m$.</p>

<p style="color: blue"><a name="Prelab2">**Pre-lab 2.**</a> When a force $\vec{F}$ is applied to an object that rotates about a fixed point at a position $\vec{r}$ relative to the fixed point, the torque produced by the force is</p>

<p style="color: blue"><a name="eq:torque">*Equation 4:*</a>  Definition of torque
\begin{equation}
  \vec{\tau} = \vec{r} \times \vec{F}.
\end{equation}</p>

<p style="color: blue">After the two large masses $M$ are introduced as indicated in [Fig. 2](#Fig2), the pendulum is allowed to come into equilibrium wherin the torque from the wire torsion balances the torque from the gravitational attraction of the lead spheres.
If $m = 10\,\mathrm{g}$, $M = 1.5\,\mathrm{kg}$, and $b = 5\,\mathrm{cm}$ as in [Pre-lab 1](#Prelab1), and if the torsion constant of the wire is $\kappa = 10^{-6}\,\mathrm{N}\,\mathrm{m}\,\mathrm{rad}^{-1}$ and the length of the arms of the pendulum is $r = 5\,\mathrm{cm}$, what is the angle $\theta_1$ between the new equilibrium and the untwisted state of the pendulum?
Assume that the angle is small enough so that $\vec{F}$ and $\vec{r}$ are approximately perpindicular.</p>

<p style="color: blue">Continuing the analogy with spring motion, Newton's Second Law for rotational motion relates the applied torque to angular acceleration, $\alpha$,</p>

<p style="color: blue"><a name="eq:newt2">*Equation 5:*</a>  Newton's Second Law for rotational motion
\begin{equation}
  \tau = I \frac{\mathrm{d}^2\theta}{\mathrm{d}t^2} \equiv I \alpha,
\end{equation}</p>

<p style="color: blue">where $I$ is the moment of inertia, which plays a role analogous to the mass in a spring system.
Assuming that the rod of the torsion pendulum is massless, for two equal masses a distance $r$ from the center of rotation,</p>

<p style="color: blue"><a name="eq:moment">*Equation 6:*</a>  Moment of inertia for torsion pendulum
\begin{equation}
  I = 2\,m\,r^2.
\end{equation}</p>

<p style="color: blue">Thus, the equation of motion for a torsion pendulum oscillating around its equilibrium is mathematically identical to the equation of motion for a spring,</p>

<p style="color: blue"><a name="eq:shm">*Equation 7:*</a>  SHM for rotational motion
\begin{equation}
  I \frac{\mathrm{d}^2\theta}{\mathrm{d}t^2} + \kappa\,\theta = 0,
\end{equation}</p>

<p style="color: blue">which has the familiar solutions of simple harmonic motion,</p>

<p style="color: blue"><a name="eq:shmsolve">*Equation 8:*</a>  Solution to SHM for rotational motion
\begin{equation}
  \theta(t) = \Theta \sin{(\omega\,t + \phi)},
\end{equation}</p>

<p style="color: blue">where $\Theta$ is the angular amplitude of the oscillation, $\omega \equiv \sqrt{\kappa/I}$ is the oscillation frequency, and $\phi$ is a constant phase.</p>

<p style="color: blue"><a name="Prelab3">**Pre-lab 3.**</a> What is the period, $T \equiv \frac{2 \pi}{\omega}$, of the torsion pendulum described in [Pre-lab 2](#Prelab2)?</p>

<p style="color: blue">Torsion pendulums can be extremely sensitive, as they must be to be affected by the gravity of an object that can fit in a lab.
The apparatus used to collect the data that we will be analyzing is sensitive enough that air resistance has an observable effect on the pendulum's oscillation.
The force of air resistance on a small moving sphere, such as the small masses $m$ at the ends of the pendulum, is approximately proportional to the velocity $\vec{v}$ of the sphere via Stokes' Law,</p>

<p style="color: blue"><a name="eq:stokes">*Equation 9:*</a>  Stokes' Law
\begin{equation}
  \vec{F}_{\mathrm{drag}} = -6\pi\,R\,\eta\,\vec{v},
\end{equation}</p>

<p style="color: blue">where $R$ is the radius of the sphere and $\eta$ is a constant of the medium called the dynamic viscosity.</p>

<p style="color: blue"><a name="Prelab4">**Pre-lab 4.**</a> Show that the torque due to air resistance can be expressed as</p>

<p style="color: blue"><a name="eq:drag">*Equation 10:*</a>  Torque due to drag
\begin{equation}
  \tau_{\mathrm{drag}} = -N \frac{\mathrm{d}\theta}{\mathrm{d}t},
\end{equation}</p>

<p style="color: blue">where $N$ is a constant.</p>

<p style="color: blue">Adding this air resistance term to the equation of motion gives the differential equation of a damped harmonic oscillator,</p>

<p style="color: blue"><a name="eq:dhm">*Equation 11:*</a>  Rotational damped harmonic oscillator
\begin{equation}
  I \frac{\mathrm{d}^2\theta}{\mathrm{d}t^2} + N \frac{\mathrm{d}\theta}{\mathrm{d}t} + \kappa\,\theta = 0,
\end{equation}</p>

<p style="color: blue">which, for an underdamped harmonic oscillator, has solutions</p>

<p style="color: blue"><a name="dhmsolve">*Equation 12:*</a> Solutions to the damped harmonic oscillator equation
\begin{equation}
  \theta(t) = \Theta_0 e^{-a t} \sin{(\omega\,t + \phi)},
\end{equation}</p>

<p style="color: blue">where $\Theta_0$ is the initial amplitude of oscillation, $a \equiv N / 2 I$, and $\omega \equiv \sqrt{\omega_0^2 - a^2}$ where $\omega_0 = \sqrt{\kappa / I}$ is the frequency of the undamped harmonic oscillator.</p>

## Part 2:  Fitting the data

<p style="color: blue">The functional form of [Eq. 14](#eq:sdhm) derived in [Part 1.1](#P1.1) is much more complicated than a polynomial.
In order to fit this function to the data, we will define our own python function to do the calculation.
If you are new to python functions, the cell below defines a short function for the motion of a simple harmonic osciallator without the damping term:</p>

```python
def shmmodel(t, S0, A0, omega, phi):
    """Compute the position of a SHM at time t."""
    return S0 + A0 * np.sin(omega * t + phi)
```

```python
## shmmodel can now be used to calculate S values for given t values
S0 = 0.4
A0 = 0.025
omega = 2 * np.pi / 1250.
phi = np.pi / 4

S = shmmodel(0.3, S0, A0, omega, phig)
print( "S = ", S, " at t = ", 0.3, " s")
```

<p style="color: blue">Some comments about the form of `shmmodel`:</p>
1. <div style="color: blue">The fitting procedure expects the independent variable, $t$ in this case, to be the first argument of the function,</div>

2. <div style="color: blue">Parameters of the model that are to be adjusted by the fitting procedure to the best representation of the data should be arguments to the function,</div>

3. <div style="color: blue">By using the NumPy version of the `sin` function, `shmmodel` is compatible with `np.array`s as input for $t$ and can be used, for example, in Matplotlib plotting (see below).</div>

<p style="color: blue">The `curve_fit` function will take as arguments the model that is supposed to define their relation (the name of your damped harmonic oscillator function), the array of values for your independent variable ($t$), and the array for the corresponding dependent variable ($S$).
It will find the values for the model parameters (_i.e._, $S_i$, $A_0$, $a$, $\omega$, and $\phi$) that minimize the $\chi^2$ between the computed model values and the measured values of $S$.
It will also compute a covariance matrix for the best-fit model parameters.</p>

There are almost always several ways to accomplish the same task in python.</p>

It does not need to be precise, but you need to start the fit procedure in the right 'ball park'.
The initial guess for the parameters should be a python list of values in the same order as the parameters appear in your function.
